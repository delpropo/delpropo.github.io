```python
import pandas as pd
import hydra
from plotnine import (
    ggplot, 
    geom_point, 
    aes, 
    labs, 
    theme, 
    element_text, 
    facet_wrap,
    theme_xkcd, 
    facet_grid,
    scale_x_date
)
    
```

```python
df = pd.read_excel("./data/2022_Spring_data.xlsx", sheet_name=None )

#sheet_names =  juggling, goals, games
juggling_df = df["juggling"]
goals_df = df["goals"]
games_df = df["games"]
 
```

```python

for unique_name in juggling_df["name"]:
    temp_mask = juggling_df["name"] == unique_name
    juggling_plot = (
        ggplot(juggling_df[temp_mask], aes(x="date", y="juggling_record"))
        + theme_xkcd()
        + geom_point(aes(color="name")) 
        #+ geom_point() 
        + theme(axis_text_x = element_text(angle = 0, vjust = 0.5, hjust=.5, size=25))
        + scale_x_date(date_labels="%m-%d-%y",date_breaks  ="1 day")
        #+ facet_wrap("name", ncol=1)
        + labs(title = " Juggling Record for " + unique_name, x="Date", y="Juggling Record")
                        )

    save_as = "juggling_" + unique_name + ".png"
    print(juggling_plot)
    juggling_plot.save(filename=save_as, path="./images", format="png")
   
```
